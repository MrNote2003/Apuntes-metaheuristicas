# Simulated Annealing (Recocido Simulado)



**¿Qué es y qué hace?**
Es un algoritmo de optimización metaheurístico inspirado en el proceso físico de la metalurgia conocido como recocido, donde un metal se calienta a altas temperaturas y se enfría lentamente para alterar sus propiedades físicas y eliminar defectos internos. Computacionalmente, su objetivo es encontrar el óptimo global de una función minimizando un "costo" o "energía". Se utiliza extensamente en el diseño de circuitos integrados, problemas de enrutamiento y calibración de modelos predictivos.

**¿Cómo funciona? (Principio, Método y Tipo)**
* **Tipo:** Algoritmo metaheurístico de búsqueda local estocástica (probabilística).
* **Método:** Exploración iterativa con probabilidad de aceptación de soluciones inferiores controlada por un parámetro de "temperatura" decreciente.
* **Principio:** 1. Se inicia con una solución aleatoria y un parámetro matemático llamado "Temperatura" en un valor muy alto.
  2. En cada iteración, se genera una modificación aleatoria a la solución actual (un "vecino").
  3. Se evalúa el vecino. Si su costo es menor (es una mejor solución), el movimiento se acepta incondicionalmente.
  4. Si su costo es mayor (es una solución peor), no se descarta de inmediato. Se utiliza una fórmula de probabilidad (basada en la distribución de Boltzmann) para decidir si se acepta. A altas temperaturas, la probabilidad de aceptar decisiones perjudiciales es muy alta, lo que permite al algoritmo saltar fuera de los "óptimos locales" y explorar el mapa completo.
  5. Progresivamente, la temperatura se reduce ("enfriamiento"). A medida que el sistema se enfría, la probabilidad de aceptar soluciones peores disminuye drásticamente, hasta que en las etapas finales el algoritmo se vuelve rígido y solo acepta movimientos que mejoren el resultado, afinando el hallazgo final.

**Ventajas y Desventajas**
* **Ventajas:** Es un algoritmo sumamente elegante y fácil de programar (requiere pocas líneas de código). Su capacidad matemática para evadir los estancamientos locales le otorga una alta probabilidad de encontrar el óptimo global sin consumir recursos de memoria masivos.
* **Desventajas:** La calidad del resultado depende críticamente de la configuración del esquema de enfriamiento. Si el proceso de enfriamiento es muy lento, la ejecución tomará un tiempo excesivo.
