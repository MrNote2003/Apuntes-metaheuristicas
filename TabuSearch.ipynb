# Tabu Search (Búsqueda Tabú)



**¿Qué es y qué hace?**
Es un algoritmo de optimización metaheurístico. Su objetivo es encontrar una solución de alta calidad en un tiempo computacional reducido, especialmente útil en problemas de gran escala donde resulta imposible calcular la solución perfecta. Se emplea frecuentemente en el ruteo masivo de vehículos, diseño de redes y planificación de horarios complejos.

**¿Cómo funciona? (Principio, Método y Tipo)**
* **Tipo:** Algoritmo metaheurístico basado en búsqueda local avanzada.
* **Método:** Exploración de vecindarios con uso de memoria a corto plazo.
* **Principio:** 1. Se parte de una solución inicial.
  2. Se analizan los "vecinos" (modificaciones menores del estado actual) y se realiza un movimiento hacia el mejor vecino disponible, **incluso si dicho movimiento empeora temporalmente la calidad de la solución**. Esto se hace para escapar de los "óptimos locales" (la falsa creencia de haber encontrado la mejor respuesta general, solo porque las opciones inmediatamente cercanas son inferiores).
  3. Para evitar que el proceso se revierta al estado anterior y caiga en un ciclo infinito, se registra el movimiento inverso en una **"Lista Tabú"** (memoria de acciones prohibidas) durante un número determinado de iteraciones.

**Ventajas y Desventajas**
* **Ventajas:** Destaca por su rapidez y eficiencia en problemas de alta complejidad. Al permitir transiciones hacia soluciones temporalmente peores, logra evadir puntos de estancamiento donde otros algoritmos quedarían atrapados.
* **Desventajas:** No se garantiza matemáticamente que la respuesta obtenida sea la óptima absoluta. Asimismo, la calibración de sus parámetros requiere un considerable proceso de experimentación.

**Parámetros y Conceptos Principales**
* solucion_inicial: El estado de partida, el cual puede generarse de forma aleatoria o mediante una heurística simple.
* lista_tabu: Estructura de memoria que almacena temporalmente los atributos de los movimientos prohibidos.
* tenencia_tabu: Define la cantidad de iteraciones que un movimiento permanece bloqueado. ¿Qué pasa si se modifica? Si el tiempo es muy corto, se olvidarán rápido las restricciones y el sistema caerá en bucles infinitos. Si es muy largo, se prohibirán demasiados caminos buenos y el algoritmo perderá la libertad necesaria para explorar zonas útiles.
* criterio_aspiracion: Es un mecanismo de excepción. ¿Qué pasa si se modifica? Permite evadir las restricciones preestablecidas.
