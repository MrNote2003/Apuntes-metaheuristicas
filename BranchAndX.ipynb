# Branch and Bound (Ramificación y Acotación)



**¿Qué es y qué hace?**
Es un algoritmo de optimización exacta para problemas combinatorios (elegir entre opciones discretas). Su trabajo es encontrar la solución óptima absoluta (la mejor posible) evaluando inteligentemente las opciones, descartando caminos inútiles desde el principio para ahorrar tiempo y memoria de procesamiento. Se usa principalmente en logística, planificación de rutas y problemas de asignación.

**¿Cómo funciona? (Principio, Método y Tipo)**
* **Tipo:** Algoritmo exacto de búsqueda en espacio de estados.
* **Método:** Divide y vencerás estructurado en un árbol.
* **Principio:**
  1. **Branch (Ramificar):** Divide el problema general en subproblemas más pequeños (crea "hijos" o ramas).
  2. **Bound (Acotar):** Calcula una "cota" (una estimación o límite matemático) de qué tan buena podría ser esa rama. Si la cota nos dice que esa rama nunca superará a una solución que ya encontramos antes, se "poda" (se descarta entera sin explorarla).

**Ventajas y Desventajas**
* **Ventajas:** Garantiza matemáticamente encontrar la mejor solución posible.
* **Desventajas:** Si la función que calcula la "cota" está mal diseñada o es débil, el algoritmo terminará revisando todas las opciones, lo que hará que el tiempo de ejecución explote (crecimiento exponencial).

**Parámetros y Conceptos Principales**
* `estado_inicial`: Los datos con los que arranca el problema (ej. matriz de distancias, capacidades).
* `funcion_cota`: La fórmula matemática que predice el futuro de una rama. Es el "cerebro" del algoritmo.
* **TSP (Problema del Agente Viajero):** Es un problema clásico donde un vendedor debe visitar un conjunto de ciudades buscando la ruta más corta y volver al inicio. Branch and Bound se usa muchísimo para resolverlo.
* **¿Qué pasa si se modifican las reglas?** Si en el TSP, por ejemplo, se bloquea una calle por reparaciones, simplemente modificas la regla de ramificación (`Branch`) para que asigne un costo infinito a esa calle. El algoritmo automáticamente podará esa rama.


```python
def branch_and_bound(estado_inicial):
    # 'mejor_solucion' guarda la respuesta final a entregar
    mejor_solucion = None
    # 'mejor_costo' inicia en infinito para que cualquier solución real lo mejore
    mejor_costo = float('inf')

    # 'nodos_pendientes' es la estructura para guardar tareas por explorar
    nodos_pendientes = [estado_inicial]

    while nodos_pendientes:
        # Se extrae un nodo para su análisis
        nodo_actual = nodos_pendientes.pop()

        # 1. BOUND (ACOTAR): Se evalúa la viabilidad del camino
        cota_actual = calcular_cota(nodo_actual)

        # Se poda la rama si la proyección no supera la solución actual
        if cota_actual >= mejor_costo:
            continue

        # 2. Se verifica la integridad de la solución
        if es_solucion_completa(nodo_actual):
            costo_real = calcular_costo(nodo_actual)

            # Se actualiza el registro histórico si se reduce el costo
            if costo_real < mejor_costo:
                mejor_solucion = nodo_actual
                mejor_costo = costo_real
        else:
            # 3. BRANCH (RAMIFICAR): Se generan los estados descendientes
            hijos = generar_subproblemas(nodo_actual)
            nodos_pendientes.extend(hijos)

    return mejor_solucion, mejor_costo

# Funciones Auxiliares

def calcular_cota(nodo):
    # Calcula y retorna una estimación numérica optimista (límite teórico) del mejor costo que podría alcanzarse explorando este nodo.
    pass

def es_solucion_completa(nodo):
    # Evalúa lógicamente si el estado del nodo representa una solución terminada que cumple con todas las reglas estructurales del problema.
    pass

def calcular_costo(nodo):
    # Procesa una solución previamente validada como completa y retorna su valor numérico exacto y definitivo.
    pass

def generar_subproblemas(nodo):
    # Ramifica el problema dividiendo el estado actual según las reglas del dominio, retornando una lista con los nuevos nodos derivados.
    pass
```

# Branch and Cut (Ramificación y Corte)



**¿Qué es y qué hace?**
Es la evolución matemática de B&B. Está diseñado específicamente para resolver modelos de **Programación Lineal Entera**. Sirve para situaciones logísticas donde la respuesta no puede tener decimales (no puedes fabricar "media silla"). Se usa en industrias pesadas, telecomunicaciones y finanzas.

**¿Cómo funciona? (Principio, Método y Tipo)**
* **Tipo:** Algoritmo exacto híbrido (Combina Branch & Bound + Planos de Corte).
* **Método:** Resolución matemática con restricciones dinámicas.
* **Principio:** Primero "relaja" el problema (finge que los decimales existen y lo resuelve con fórmulas rápidas). Si el resultado da un decimal (ejemplo, X = 3.5), usa geometría matemática para crear un "Corte" (una nueva regla que rebana el 3.5 sin eliminar los números enteros válidos como el 3 o el 4). Si ya no puede cortar más, ramifica.

**Ventajas y Desventajas**
* **Ventajas:** Es brutalmente eficiente. Reduce un árbol de millones de ramas a solo unas pocas al cortar áreas enteras de respuestas inválidas mediante matemática pura.
* **Desventajas:** Requiere librerías matemáticas avanzadas (solvers como Gurobi o PuLP). Programar los "Cortes" a mano requiere un nivel altísimo de álgebra.

**Parámetros y Conceptos Principales**
* `modelo_matematico`: Representa el problema traducido a ecuaciones algebraicas.
* `generador_cortes`: El algoritmo secundario que busca cómo rebanar los decimales.
* **Plano de Corte:** Es una restricción temporal que se inventa el algoritmo en tiempo de ejecución para achicar el área de búsqueda.
* **¿Qué pasa si se modifican las reglas?** Si las variables de repente sí pueden ser fraccionarias (ej. en lugar de camiones son litros de agua), entonces no necesitas Branch and Cut. Eliminas los cortes y simplemente lo resuelves como un problema de programación lineal estándar en milisegundos.

# Branch and Price (Ramificación y Generación de Columnas)



**¿Qué es y qué hace?**
Es el "hermano mayor" diseñado para problemas de escala gigantesca. Se usa cuando el problema tiene millones o miles de millones de variables posibles (llamadas "columnas"), tantas que la memoria RAM de tu computadora colapsaría si intentas cargarlas todas al mismo tiempo. Es el rey indiscutido en la creación de itinerarios de aerolíneas o distribución masiva.

**¿Cómo funciona? (Principio, Método y Tipo)**
* **Tipo:** Algoritmo exacto híbrido (Branch & Bound + Generación de Columnas).
* **Método:** Problema Maestro Restringido + Subproblema de Pricing.
* **Principio:** Arranca con un modelo casi vacío, con unas poquitas variables válidas. Luego, lanza un pequeño algoritmo explorador (Pricing) que analiza las matemáticas del modelo y se pregunta: *"Si invento una variable nueva ahora mismo, ¿mejorará mi costo general?"*. Si es así, la crea, la inyecta al modelo, y recalcula. Solo ramifica cuando ya no hay más variables útiles que inventar.

**Ventajas y Desventajas**
* **Ventajas:** Permite resolver problemas titánicos que literalmente son imposibles de cargar en memoria por otros medios.
* **Desventajas:** Extremadamente complejo de codificar. La sincronización entre el problema principal y el creador de columnas requiere una arquitectura de software impecable.

**Parámetros y Conceptos Principales**
* `modelo_maestro`: El contenedor principal que guarda solo las variables activas útiles.
* `problema_pricing`: El sub-algoritmo que funciona como "cazador de variables".
* **Columna:** Representa un patrón de decisión completo (Ej. Un turno entero de trabajo de un guardia de lunes a viernes).
* **¿Qué pasa si se modifican las reglas?** Si cambian las restricciones operativas (ej. por ley, los conductores no pueden manejar más de 5 horas seguidas), modificas únicamente el `problema_pricing` para que jamás genere ni proponga variables que rompan esa regla legal.
