# Ant Colony Optimization (Optimización por Colonia de Hormigas)



**¿Qué es y qué hace?**
Es un algoritmo de optimización metaheurístico inspirado en la naturaleza, específicamente en el comportamiento de búsqueda de alimento de las hormigas reales. Su objetivo es encontrar rutas óptimas en gráficos o redes complejas. Se emplea ampliamente en problemas de enrutamiento de vehículos, diseño de redes de telecomunicaciones y logística de transporte dinámico, donde las condiciones pueden cambiar en tiempo real.

**¿Cómo funciona? (Principio, Método y Tipo)**
* **Tipo:** Algoritmo metaheurístico basado en Inteligencia de Enjambre (Swarm Intelligence).
* **Método:** Construcción probabilística de soluciones mediante rastros de feromonas artificiales.
* **Principio:** 1. Múltiples agentes artificiales ("hormigas") construyen rutas paso a paso a través de una red. La decisión de qué camino tomar se basa en la probabilidad, influenciada por qué tan cerca está el siguiente punto y cuánta feromona tiene ese camino.
  2. Una vez que todas las hormigas terminan sus recorridos, se evalúa la calidad de cada ruta (por ejemplo, la distancia total).
  3. Se actualizan los rastros de feromonas en el mapa: se deposita más feromona en los caminos que resultaron ser más cortos y eficientes. Simultáneamente, las feromonas de todos los caminos sufren una "evaporación" gradual. Esto provoca que las rutas ineficientes pierdan su rastro con el tiempo y las rutas óptimas se refuercen, atrayendo a más hormigas en la siguiente iteración.

**Ventajas y Desventajas**
* **Ventajas:** Posee una excelente capacidad para adaptarse a cambios dinámicos (si un camino se bloquea de repente, la evaporación y la exploración natural de las hormigas encontrarán rápidamente una ruta alternativa).
* **Desventajas:** Su convergencia (el tiempo que tarda en decidir cuál es la solución final) puede ser más lenta que la de otros métodos. Además, la matemática detrás de la actualización de probabilidades consume considerables recursos computacionales.

**Parámetros y Conceptos Principales**
* numero_hormigas: La cantidad de agentes que exploran el mapa simultáneamente en cada ronda.
* tasa_evaporacion: El porcentaje de feromona que desaparece en cada iteración. ¿Qué pasa si se modifica? Si es muy alta, se olvidarán demasiado rápido las rutas buenas y la búsqueda se volverá errática y aleatoria. Si es muy baja, se acumulará rastro en caminos ineficientes y el sistema convergerá prematuramente hacia una solución subóptima.
* alfa (peso de la feromona) y beta (peso de la visibilidad/distancia): Controlan la toma de decisiones. ¿Qué pasa si se modifican? Si el parámetro alfa es excesivamente alto, todas las hormigas seguirán ciegamente el primer rastro encontrado, limitando la exploración. Si el parámetro beta es excesivo, el algoritmo se volverá "miope", eligiendo siempre el camino inmediatamente más corto sin considerar el costo a largo plazo.
