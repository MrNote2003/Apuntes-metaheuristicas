# Genetic Algorithm (Algoritmo Genético)



**¿Qué es y qué hace?**
Es un algoritmo de optimización y búsqueda metaheurístico inspirado en los principios de la evolución biológica y la selección natural de Charles Darwin. Su objetivo es evolucionar progresivamente un conjunto de posibles soluciones hasta encontrar una de alta calidad. Se utiliza de forma extensa en el entrenamiento de modelos de Machine Learning, diseño automatizado de piezas de ingeniería y en problemas donde el espacio de búsqueda es inabarcable e irregular.



**¿Cómo funciona? (Principio, Método y Tipo)**
* **Tipo:** Algoritmo metaheurístico poblacional (Algoritmo Evolutivo).
* **Método:** Búsqueda estocástica basada en selección natural, cruzamiento y mutación genética.
* **Principio:** 1. Se genera una "población" inicial de soluciones aleatorias (donde cada solución es un "individuo" o "cromosoma").
  2. Se evalúa a cada individuo utilizando una **Función de Aptitud (Fitness)** para determinar qué tan bueno es resolviendo el problema.
  3. Se seleccionan los individuos más aptos ("supervivencia del más fuerte") para que actúen como padres.
  4. Se aplica un **Cruzamiento (Crossover)**, mezclando la "genética" (los datos) de dos padres para crear descendencia con características combinadas.
  5. Se aplica una **Mutación** aleatoria a algunos descendientes para inyectar material genético nuevo y mantener la diversidad.
  6. La nueva generación reemplaza a la antigua y el ciclo se repite hasta alcanzar un número máximo de generaciones o una solución satisfactoria.

**Ventajas y Desventajas**
* **Ventajas:** Al procesar múltiples soluciones a la vez (una población completa), puede explorar muchas áreas distintas del problema de forma simultánea. Es altamente paralelizable y no requiere que el problema tenga una formulación matemática estricta.
* **Desventajas:** Computacionalmente costoso, ya que se debe calcular el desempeño de cada individuo en cada generación. No asegura la obtención del óptimo absoluto.

**Parámetros y Conceptos Principales**
* `tamaño_poblacion`: La cantidad de soluciones (individuos) que existen en cada generación. **¿Qué pasa si se modifica?** Si es muy pequeña, la diversidad genética será pobre y el algoritmo se estancará rápidamente en una solución mediocre. Si es excesivamente grande, el tiempo de procesamiento por cada iteración será inmanejable.
* `tasa_cruzamiento`: La probabilidad de que dos padres intercambien su información para crear un hijo. **¿Qué pasa si se modifica?** Si es demasiado baja, el algoritmo operará como una simple búsqueda aleatoria en lugar de combinar las características exitosas descubiertas.
* `tasa_mutacion`: La probabilidad de que un "gen" (una variable) cambie su valor al azar. **¿Qué pasa si se modifica?** Si es muy baja, la población se volverá idéntica y no se descubrirán nuevas áreas del problema (pérdida de diversidad). Si es muy alta, el algoritmo destruirá las combinaciones buenas constantemente, transformándose en una exploración caótica sin rumbo.
* `funcion_aptitud`: La métrica de puntuación que clasifica el desempeño de cada solución.
